```python
data2 = df.groupby("Continent")['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by = '2022 Population', ascending = False)
data2
```

```python
data3.plot()

#The population behavior over the years shown using a line graph.
#The Asian population looks to explode in numbers between 1990 till 2010, While the other continents maintain 
#a gradual increase thoughout the years. Europe is the only one that maintains a plateau.
```
